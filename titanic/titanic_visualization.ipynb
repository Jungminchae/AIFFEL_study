```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
pd.set_option('display.max_rows', 100)
```

```python
# Data 불러오기
train_path = os.path.join('data', 'train.csv')
test_path =os.path.join('data', 'test.csv')
sub_path = os.path.join('data', 'gender_submission.csv')

train = pd.read_csv(train_path)
test = pd.read_csv(test_path)
sub = pd.read_csv(sub_path)
```

```python
train.head()
```

```python
# sns.countplot()
for col in ['Survived', 'Pclass', 'Sex', 'SibSp', 'Parch', 'Embarked']:
    print('col : {} \n'.format(col))
    print('-'*50)
    
    sns.countplot(train[col])
    plt.show()
```

```python
# countplot, 내림차순으로 정렬해서 그리기
for col in ['Survived', 'Pclass', 'Sex', 'SibSp', 'Parch', 'Embarked']:
    print('col : {} \n'.format(col))
    print('-'*50)
    
    sns.countplot(train[col], order=train[col].value_counts().sort_values(ascending=False).index)
    plt.show()
```

```python
# 연속형 변수 분포확인 sns.displot
for col in ['Age','Fare']:
    print('col : {} \n'.format(col))
    print('-'*50)
    
    sns.distplot(train[col])
    plt.show()
```

```python
# 연속형 변수 분포확인 sns.displot
# kde = False histogram만 가능 
# bins로 막대수 조정 가능
for col in ['Age','Fare']:
    print('col : {} \n'.format(col))
    print('-'*50)
    
    sns.distplot(train[col], kde=False, bins=30)
    plt.show()
```

```python
# numerical 데이터 타입 분포확인
sns.pairplot(train.drop('PassengerId', axis=1))
```

```python
# Corr
cor = train.drop('PassengerId',axis=1).corr()
```

```python
# heatmap
plt.figure(figsize=(10,10))
sns.heatmap(cor)
```

```python
# 생존자 중에 성별 확인
# 여성이 훨씬 생존률이 높음
sns.countplot(train['Survived'] , hue=train['Sex'])
```

```python
# 생존자 중에 Pclass 확인
# pclass가 3인 경우 생존률 매우 떨어짐
sns.countplot(train['Survived'], hue=train['Pclass'])
```
